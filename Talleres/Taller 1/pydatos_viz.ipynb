### Plot usando un diccionario de datos

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(42)

data = {'edad': np.arange(50)}
print(data['edad'])

data['estatura'] = data['edad'] + 10 * np.random.randn(50)
print(data['estatura'])

fig, ax = plt.subplots()
ax.scatter('edad', 'estatura', data=data)
ax.set(xlabel='edad', ylabel='estatura')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(40)

data = {'x': np.arange(50)}
data['y'] = data['x'] + 10 * np.random.randn(50)
data['pob'] = np.random.randint(0, 50, 50)

fig, ax = plt.subplots()
ax.scatter('x', 'y', c='pob', data=data, s='pob') #s: size, c:color
ax.set(xlabel='x', ylabel='y')
plt.show()
```

### Cambiando el mapa de colores

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(40)

data = {'x': np.arange(50)}
data['y'] = data['x'] + 10 * np.random.randn(50)
data['pob'] = np.random.randint(0, 50, 50)

fig, ax = plt.subplots()
ax.scatter('x', 'y', c='pob', data=data, cmap='magma') #s: size, c:color
ax.set(xlabel='x', ylabel='y')
plt.show()
```

### Scatter - Dispersión

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(40)

n = 50
x = np.random.rand(n)
y = np.random.rand(n)
z = x**2
w = np.pi*np.random.randint(0,10,n)**2

plt.scatter(x, y, c=z) #c: color
plt.scatter(x, x, s=w, alpha=0.6) # s: tamaño/area - alpha: transparencia
plt.show()
```

### Barras

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

productos = ['mesas', 'sillas', 'repisas', 'camas']
inventario = [30, 80, 50, 40]
colores = ['tab:red', 'tab:blue', 'tab:orange', 'tab:red']

ax.bar(productos, inventario, color=colores)

ax.set_ylabel('Nivel de inventario')
ax.set_title('Inventario de productos')

plt.show()
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

productos = ['mesas', 'sillas', 'repisas', 'camas']
inventario = [30, 80, 50, 40]
colores = ['tab:red', 'tab:blue', 'tab:orange', 'tab:red']

ax.barh(productos, inventario, color=colores)

ax.set_xlabel('Nivel de inventario')
ax.set_title('Inventario de productos')

plt.show()
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

productos = ['mesas', 'sillas', 'repisas', 'camas']
inventario = [30, 80, 50, 40]
colores = ['tab:grey', 'tab:blue', 'tab:grey', 'tab:grey']

bar_pos = [0,0.4,0.8,1.2]
ax.barh(bar_pos, inventario, color=colores, height=0.3)

ax.set_yticks(bar_pos, productos)
ax.set_xlabel('Nivel de inventario')
ax.set_title('Inventario de productos')

plt.show()
```

```python
import matplotlib.pyplot as plt

productos = ['mesas', 'sillas', 'repisas', 'camas']
inventario = [30, 80, 50, 40]

explode = (0.1, 0, 0, 0)  # separa la primera parte
cmap = plt.colormaps["tab20c"]

fig, ax = plt.subplots()
ax.pie(inventario, labels=productos, autopct='%d%%', explode=explode, startangle=0,colors=cmap([0,1,5,6]))
plt.show()
```

```python
import matplotlib.pyplot as plt

productos = ['mesas', 'sillas', 'repisas', 'camas']
inventario = [30, 80, 50, 40]

explode = (0.1, 0, 0, 0)  # separa la primera parte
cmap = plt.colormaps["tab20c"]

fig, ax = plt.subplots()
ax.pie(inventario, labels=productos, colors=cmap([0,1,5,6]), wedgeprops=dict(width=0.5))
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(40)
fig, ax = plt.subplots()
ax.plot(np.random.standard_normal(1000).cumsum())
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(40)
fig, ax = plt.subplots()
ax.plot(np.random.standard_normal(1000).cumsum())
ax.set_xticks(range(0,1100,100))
ax.set_xticklabels(range(0,51,5), rotation=30, fontsize=10)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(40)

fig, ax = plt.subplots()

ax.plot(np.random.standard_normal(1000).cumsum())

ax.set_xticks(range(0,1000,365))
ax.set_xticklabels(['Año 1', 'Año 2', 'Año 3'], rotation=0, fontsize=12)

for label in ax.get_xticklabels():
    label.set_horizontalalignment('left')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(40)
fig, ax = plt.subplots()
ax.plot(np.random.standard_normal(1000).cumsum(),label="TLA")
ax.plot(np.random.standard_normal(1000).cumsum(),label="APL")
ax.plot(np.random.standard_normal(1000).cumsum(),label="MSF")
ax.legend() #loc="upper right"
plt.show()
```

```python
spx_df = pd.read_csv("spx.csv",index_col=0, parse_dates=True)
```

```python
from datetime import datetime

fig, ax = plt.subplots()

spx = spx_df["SPX"]

spx.plot(ax=ax, color="black")
```

```python
# Zoom en 2007-2010
ax.set_xlim(["1/1/2007", "1/1/2011"])
ax.set_ylim([600, 1800])
fig
```

Anotaciones

```python
crisis_data = [
    (datetime(2007, 10, 11), "Peak of bull market", 75),
    (datetime(2008, 3, 12), "Bear Stearns Fails", 80),
    (datetime(2008, 9, 15), "Lehman Bankruptcy", 85)
]

for date, label, dist in crisis_data:
    ax.annotate(label, xy=(date, spx.asof(date) + dist),
                xytext=(date, spx.asof(date) + 225),
                arrowprops=dict(facecolor="black", headwidth=4, width=2,
                                headlength=4),
                horizontalalignment="left", verticalalignment="top")


ax.set_title("Important dates in the 2008–2009 financial crisis")
fig
```

```python
import matplotlib.pyplot as plt
plt.close()
fig, ax = plt.subplots()
spx.plot(ax=ax, color="blue")

ax.set_xlim(["1/1/2007", "1/1/2011"])
ax.set_ylim([600, 1800])

crisis_data = [
    (datetime(2007, 10, 11), "Peak of bull market"),
    (datetime(2008, 3, 12), "Bear Stearns Fails"),
    (datetime(2008, 9, 15), "Lehman Bankruptcy")
]

for date, label in crisis_data:
    ax.annotate(label, xy=(date, spx.asof(date) +70),
                xytext=(date, spx.asof(date) + 225),
                arrowprops=dict(facecolor="black", headwidth=4, width=1,
                                headlength=4),
                horizontalalignment="left", verticalalignment="top")


ax.set_title("Important dates in the 2008–2009 financial crisis")
plt.show()
```
